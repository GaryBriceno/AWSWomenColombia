```python
# Una cadena en Python es una secuencia de caracteres

palabra = 'Python'

for caracter in palabra:
  print(caracter)
```

```python
# Por defecto se agrega una nueva linea al final de cada caracter
# Es decir, el valor por defecto de "end" es una nueva linea

for caracter in palabra:
  print(caracter, end=' ')
```

```python
# Utilizaremos el for, que nos permitira iterar sobre una coleccion de elementos

python_libraries = ['numpy', 'pytorch', 'spacy', 'pandas', 'seaborn']

for library in python_libraries:
  print(library)
```

```python
# vamos a iterar en una secuencia
# range(5), genera una secuencia de numeros enteros de 0 a 4

for counter in range(5):
  print(counter, end=", ")
```

```python
# Iteracion en forma descendente

for counter in range(10, 0, -1):
  print(counter, end = ', ')
```

```python
# recorrer set
my_set = {1, 3, 4, 4}
for element in my_set:
  print(element)
```

```python
my_tupla = (1, 3, 4)
for element in my_tupla:
  print(element)
```

```python
my_dictionary = {"Argentina":"ar", "Colombia":"co", "Peru":"pe" }
for key in my_dictionary:
  print(key)
```

```python
for item in my_dictionary.items():
  print(item)
```

```python
for key in my_dictionary.keys():
  print(key)
```

```python
for value in my_dictionary.values():
  print(value)
```
