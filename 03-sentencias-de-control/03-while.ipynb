## Uso del bucle While

```python
contador = 0

while contador < 5:
  print(contador)
  contador += 1
```

```python
# El while se utiliza generalmente cuando no se conoce la cantidad de iteracciones a realizar
# mas si se conoce la condicion de verdad que hace que el bucle termine

sentinel = 0

while sentinel != -1:
  print("Existen aun lineas por leer")
  sentinel = int(input("Ingrese un valor: "))
```
