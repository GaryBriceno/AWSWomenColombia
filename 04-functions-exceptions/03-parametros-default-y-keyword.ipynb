```python
# Se pueden proporcionar parametros por defecto
def rectangle_area(length=2, width=3):
  """ Return a rectangle's area """
  return length * width
```

```python
# Se utilizan ambos valores por defecto
rectangle_area()
```

```python
# Se utiliza el valor de width por defecto
rectangle_area(3)
```

```python
# No se utiliza ningun valor por defecto
rectangle_area(4, 5)
```

```python
# Tambien se pueden usar los parametros, utilizando el nombre
def rectangle_area_keword(length, width):
  """ Return a rectangle's area """
  return length * width
```

```python
rectangle_area_keword(length=4, width=20)
```

```python
rectangle_area_keword(width=20, length=5)
```

```python
# El uso de args como nombre es una convencion
def average(*args):
  return sum(args)/len(args)
```

```python
# Esto indica a Python: agrupa el listado de parametros como una tupla y pasa dicha tupla como parametro
average(4, 5, 6)
```

```python
average(4, 5, 6, 10)
```

```python
average()
```

```python
# Esto se puede utilizar en funciones con argumentos tambien
grades = [88, 75, 96, 55, 83]
```

```python
# De esta forma, al utilizar * al lado de la variable, se esta 
# desempacando los valores del listado y pasando a la funcion
average(*grades)
```

```python
# Queremos hacer una consulta a una base de datos
# utilizamos algunos filtros
def filter(ciudad, provincia, fecha_alta):
    return "SELECT * FROM clientes " 
           "WHERE ciudad='{}' AND provincia='{}' AND fecha_alta={};".format(ciudad, provincia, fecha_alta)
```

```python
def filter(**kwargs):
    query = "SELECT * FROM clientes"
    i = 0
    for key, value in kwargs.items():
        if i == 0:
            query += " WHERE "
        else:
            query += " AND "
        query += "{}='{}'".format(key, value)
        i += 1
    query += ";"
    return query
```

```python
# El orden de los parámetros *args y **kwargs en la definición de una función importa, y mucho. 
# Ambos pueden aparecer de forma conjunta o individual, 
# pero siempre al final y de la siguiente manera:
def ejemplo(arg1, arg2, *args, **kwargs)
```
