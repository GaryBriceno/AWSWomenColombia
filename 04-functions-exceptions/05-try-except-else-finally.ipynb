## Manejo de Exceptions o Excepciones



```python
divisor = 0
numeral = 10
```

```python
division = numeral / divisor
```

```python
try:
  division = numeral/divisor
except:
  print("Existe un error en la division")
```

```python
try:
  division = numeral / divisor
except ZeroDivisionError:
  print("Error de division por cero")
except:
  print("Algun error adicional")
```

```python
try:
  raise ValueError()
  division = numeral / divisor
except ZeroDivisionError:
  print("Error de division por cero")
except:
  print("Algun error adicional")
```

```python
try:
  raise ValueError("Existe un error previo a la division")
  division = numeral / divisor
except ZeroDivisionError:
  print("Error de division por cero")
except Exception as e:
  print(str(e))
```

```python
try:
  raise ValueError("Existe un error previo a la division")
  division = numeral / divisor
except ZeroDivisionError:
  print("Error de division por cero")
except Exception as e:
  print(str(e))
finally:
  print("Bloque de codigo que se ejecuta, si es agregado")
```

```python
try:
  divisor = 0
  division = numeral/divisor
except:
  print("Se presento una Exception")
else:
  print("Codigo que se ejecuta si no hay exception")
finally:
  print("bloque finally")
```

```python
try:
  divisor = 1
  division = numeral/divisor
except:
  print("Se presento una Exception")
else:
  print("Codigo que se ejecuta si no hay exception")
finally:
  print("bloque finally")
```
