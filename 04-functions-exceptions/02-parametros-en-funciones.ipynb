## Parametros en Funciones

* Pasar datos entre funciones
  - En Python todos son objetos y los valores se pasan por referencia
* Retornar multiples valores en una funcion
* Alcance de las funciones


```python
# Definicion de una funcion con muchos parametros
# Ingresamos los parametros separados por coma

def maximum(value1, value2, value3):
  # La primera parte de una funcion es el docstring
  """ Return the maximum of three values."""
  max_value = value1
  if value2 > max_value:
    max_value = value2
  if value3 > max_value:
    max_value = value3
  return max_value
```

```python
# La funcion es un objeto y el docstring es data parte de la funcion
maximum(12, 27, 36)
```

```python
maximum(12.3, 45.6, 9.7)
```

```python
# El operador > tambien funciona con cadenas
maximum('yellow','red','orange')
```

```python
# Se puede utilizar la funcion con diferentes valores, que son compatibles
# Se puede mezclar tipos en algunos escenarios
maximum(13.5, -3, 7)
```

```python
# Siempre revisar si la funcion que necesitan esta implementada
max('yellow', 'red', 'orange', 'blue', 'green')
```

```python
min(15, 9, 27, 14)
```

```python
listado = [15, 9, 27, 14]
max(listado)
```

```python
# Se utiliza la palabra como una secuencia de caracteres
min('orange')
```

```python
# Existe una forma de generar la misma secuencia
random.seed(32)

for roll in range(10):
  print(random.randrange(1, 7), end=' ')
```

```python
# Los valores son similares a los que se generan la primera vez
# Esto sirve para simular funciones, cuando se necesita el mismo
# conjunto de valores
random.seed(32)
for roll in range(10):
  print(random.randrange(1, 7), end=' ')
```
