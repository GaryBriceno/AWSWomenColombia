```python
# Sintaxis basica de una funcion
# Inicia con la palabra reservada def
# Se indica el nombre de la funcion
# Entre parentesis se indica los parametros
# En python el indextado es IMPORTANTE

def square(number):
  # El texto entre tres comillas es docstring o documentacion de la funcion
  """ Calculate the square of number. """
  return number ** 2
```

```python
# Esto es algo que se puede ejecutar en multiples ocasioens
square(4)
```

```python
variable = square(10)
variable
```

```python
# En otros lenguajes, se debe indicar el tipo y la funcion solo puede recibir 
# el tipo especificado

# En el caso de Python se puede ingresar tipos diferentes que cumplan con la funcion
variable_float = square(2.5)
variable_float
```

```python
# Se puede pasar toda aquel tipo que soporte el operador **
other_variable = square("text")
other_variable
```

```python
# En Python siempre se devuelve un valor, cuando se llama a la funcion
# Ya sea que exista la palabra return o no

# Si no esta la palabra return, Python retorna la palabra return

def function_without_return(nombre):
  print("hola", nombre)

my_function = function_without_return("Gary")
print(my_function)
```
