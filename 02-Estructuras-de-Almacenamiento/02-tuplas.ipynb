## Tuplas

* Definición de una tupla
* Operaciones con tuplas
* Inmutabilidad en las tuplas
* Uso de tuplas


```python
# Definicion de una tupla
some_tuple = (1, 2, 3)
```

```python
# Acceder a un elemento de la tupla
some_tuple[1]
```

```python
# Se pueden realizar busquedas en una tupla
0 in some_tuple
```

```python
# Una tupla es inmutable
some_tuple[1] = 5
```

```python
# Sin embargo, si podemos modificar elementos mutables dentro de ella
another_tuple = ( 1, [2, 3], 4)
another_tuple
```

```python
another_tuple[1]
```

```python
another_tuple[1].append(5)
another_tuple
```

```python
# Al ser inmutable, se puede utilizar para guardar constantes
weekdays = ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun')
```
