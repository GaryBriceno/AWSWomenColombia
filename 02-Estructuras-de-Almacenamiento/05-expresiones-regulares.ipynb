## Regular Expressions

Se utilizaban inicialmente en sistemas como UNIX.

Son patrones que definen la forma que un texto debe tener.
- Ejemplos: codigo postal, emails, telefonos

Se puede utilizar para:     
- Validación de datos
- Limpieza de data
- Transformación de datos

Repositorios que pueden consultar:  
- https://regex101.com/
- https://regexlib.com/
- https://www.regular-expressions.info/

```python
import re
```

```python
pattern = '02215'
print(re.fullmatch(pattern, '022'))
```

```python
print(re.fullmatch(pattern, '02215'))
```

```python
# Validar si mi patron tiene cinco digitos
re.fullmatch(r'\d{5}', '02215')
```

```python
print(re.fullmatch(r'\d{5}', '022154'))
```

Se tiene: 

* \d - cualquier digito entre (0 - 9)
* \D - cualquier caracter que no es un digito
* \s - cualquier caracter que sea un espacio en blanco
* \S - cualquier caracter que no sea un espacio en blanco
* \w - cualquier caracter alfanumerico
* \W - cualquiera que no sea un caracter alfanumerico

```python
print(re.fullmatch(r'\D{4}', '1234'))
```

```python
print(re.fullmatch(r'\D{4}', 'abcd'))
```

```python
print(re.fullmatch('[A-z][a-z]*', 'Wally'))
```
