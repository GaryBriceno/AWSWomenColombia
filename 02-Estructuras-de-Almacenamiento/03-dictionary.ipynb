## Dictionaries

* Definicion de diccionarios
* Creacion, inicialiacion y consulta de elementos
* Agregar, eliminar y actualizar elementos
* Operadores para comparar diccionarios

### Caracteristicas

* Son mutables
* Son dinámicos: Se pueden agregar o quitar valores 
* Se pueden anidar 
* Se accede a los elementos mediante las llaves

```python
# Conjunto de elementos llave-valor

country_codes = {'Colombia':'co', 'Peru':'pe', 'Ecuador':'ec', 'Argentina':'ar'}
country_codes
```

```python
# Las llaves no se pueden repetir
country = {'Colombia':'co', 'Peru':'pe', 'Ecuador':'ec', "Peru":"ar"}
country
```

```python
len(country_codes)
```

```python
# Creacion de un diccionario
person = {}
person["name"] = 'Joe'
person["last_name"] = 'Smith'
person['children'] = ['Ralph', 'Betty', 'Joey']
person['pets'] = {'dog':'Fido', 'cat': 'Sox'}

person
```

```python
roman_numerals = {'I':1, 'II':2, 'III':3, 'V':5, 'X':100} 
roman_numerals
```

```python
# Se puede acceder al elemento utilizando la Llave
roman_numerals['V']
```

```python
# Los diccionarios son mutables, se pueden actualizar los valores
roman_numerals['X'] = 10
roman_numerals
```

```python
# Si el valor de la llave no existe, se agrega un nuevo valor
roman_numerals['L'] = 50
roman_numerals
```

```python
# Eliminar un elemento
del roman_numerals['L']
roman_numerals
```

```python
# Tambien se puede utilizar el metodo "pop" para eliminar un elemento
roman_numerals.pop('X')
roman_numerals
```

```python
# Acceder a los elementos
roman_numerals['X']
```

```python
print(roman_numerals.get('X'))
```

```python
roman_numerals.get('V')
```

```python
# Se puede indicar que se imprima un valor alternativo cuando no exista la llave
print(roman_numerals.get('X', 'El valor no existe en el diccionario'))
```

```python
# Validar si una llave se encuentra en un diccionario
'V' in roman_numerals
```

```python
roman_numerals.keys()
```

```python
roman_numerals.items()
```

```python
roman_numerals.values()
```
