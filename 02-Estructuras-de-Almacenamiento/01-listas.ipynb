## Listas

* Definición de una lista
* Operaciones con listas
* Conversión de cadenas en listas
* Uso de la función range



```python
# Una lista inicia con corchetes o brackets
empty_list = []
print(empty_list)
```

```python
# Los elementos de la lista se separan por comas
python_libraries = ['pandas', 'numpy', 'spacy', 'scikit-learn']
print(python_libraries)
```

```python
# Una lista vacia es como si fuera False
if empty_list:
  print("empty_list is True")
else:
  print("empty_list is False")
```

```python
# Una lista con elementos es True
if python_libraries:
  print("List with elements is True")
```

```python
# Se puede acceder a cada elemento de la lista
# El contador inicia con cero
print(python_libraries[0])
```

```python
id(python_libraries)
```

```python
# El contenido puede cambiar.
# Una lista por definicion es "mutable"
python_libraries[1] = "math"
print(python_libraries)
```

```python
id(python_libraries)
```

```python
# Se pueden agregar elementos a la lista

python_libraries.append("TensorFlow")
print(python_libraries)
```

```python
# Una palabra se puede convertir en una lista
word = "Python"
word_list = list(word)
print(word_list)
```

```python
# Se puede acceder al ultimo elemento
python_libraries[-1]
```

```python
# Se puede acceder a un rango de la lista
python_libraries[1:3]
```

```python
python_libraries[:3]
```

```python
python_libraries[-2:]
```

```python
# Se puede devolver una lista indicando un salto
python_libraries[::2]
```

```python
# Tambien se puede retornar los elementos en orden descendente de indices
python_libraries[::-1]
```

```python
# Una lista puede contener cualquier tipo de elementos
different_elements = [1, 'Juan', True]
different_elements
```

```python
# Una lista puede contener otra lista
list_one = ["Peru", 'Colombia', 'Ecuador']
list_two = ['Africa', 'Europa', list_one]
list_two
```

```python
# La function range
list(range(10))
```

```python
list(range(1,10))
```

```python
# Tambien se puede empezar por un numero negativo
list(range(-10,5))
```

```python
# Tambien se puede proporcionar un incremento para la lista, en este caso 2
list(range(1, 20, 2))
```

```python
# Solo acepta como parametros enteros
list(range(1.2))
```

```python
elements = list(range(10))
elements
```

```python
elements.insert(2, 18)
elements
```

```python
# La funcion pop() sirve para eliminar el ultimo elemento de la lista
elements.pop()
elements
```

```python
del elements[0]
elements
```

```python
# Se pueden combiar listas
caribe = ['Cuba', 'Puerto Rico', 'Republica Dominicana']
europa = ['Francia', 'Italia', 'Polonia']
paises = caribe + europa
paises
```

```python
# Buscar en una lista
'Cuba' in paises
```

```python
# Posicion de un item
paises.index('Polonia')
```
