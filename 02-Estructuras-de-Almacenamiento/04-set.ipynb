## Set

* Definicion de un set
* Operaciones con el set
* Comparación de set

```python
# Set es una colección de valores únicos
set_values = {1, 2, 3, 4, 4, 4}
set_values
```

```python
# Los elementos pueden ser de diferentes tipos
colors = {"red", "blue", 1}
colors
```

```python
# Se puede determinar el numero de elementos
len(colors)
```

```python
"red" in colors
```

```python
"purple" in colors
```

```python
numbers = list(range(10)) + list(range(5))
numbers
```

```python
unique_numbers = set(numbers)
unique_numbers
```

```python
# Como crear un set de datos vacios
empty_set = {}
empty_set
```

```python
type(empty_set)
```

```python
real_empty_set = set()
type(real_empty_set)
```

```python
# En base a una cadena, obtener las palabras unicas y luego presentarlos en forma ordenada
text = "to be or not to be that is the question"

text.split()
```

```python
unique_words = set(text.split())
unique_words
```

```python
sorted(unique_words)
```

```python
# Se compara si todos los elementos de un set se encuentran en el otro
{1, 2, 3} == {2, 3, 1}
```

```python
{1, 2, 3, 4} == {2, 3, 4}
```

```python
# Validar si un set esta contenido en otro.
{1, 2} < {3, 4}
```

```python
{1, 2} < {1, 2, 3, 4}
```

```python
{1, 2}.issubset({1, 2, 3, 4})
```

```python
{1, 2, 3, 4}.issuperset({1, 2})
```

```python
# Union de dos set
{1, 2, 3} | {3, 4, 5}
```

```python
{1, 2, 3}.union([3, 4, 5])
```

```python
# Interseccion de elementos
{1, 3, 5} & {2, 3, 4}
```

```python
{1, 3, 5}.intersection([2, 3, 4])
```

```python
# Diferencia de elementos
{1, 2, 3, 4 } - {3, 4}
```

```python
# Diferencia simetrica
{1, 3, 5} ^ { 1, 3, 4}
```

```python
{1, 3, 5}.symmetric_difference({1, 3, 4})
```

```python
# Valida si no tiene elementos en comun
{1, 3, 5}.isdisjoint({2, 3, 6})
```

```python
{1, 3, 5}.isdisjoint({2, 4, 6})
```
