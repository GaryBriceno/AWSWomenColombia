## String

* Operador +
* Operador *
* Uso del in y not in
* funciones de cadena



```python
# El operador +, concatena cadenas
s = "foo"
t = "bar"
u = "baz"

s + t + u
```

```python
# El operador *, multiplica la cadena
3 * s
```

```python
# Si es negativo, da una cadena vacia
-1 * s
```

```python
# El uso de "in"
"f" in s
```

```python
# El uso del "not in"
"r" not in s
```

```python
s = "Esto es una cadena"
len(s)
```

```python
# Indice en una cadena
s[0]
```

```python
s[0:5]
```

```python
# Agregando un salto
s[::2]
```

```python
# Una cadena o string es inmutable
name = "Juana"
name[1] = "u"
```

```python
id(name)
```

```python
print(name.replace("J","R"))
name
```

```python
id(name)
```
