## Asignación Dinámica

```python
# En python todo valor es un objeto
# La asignacion del tipo es en forma dinamica

# String variable = "Esto es una asignacion en Java"

variable = "Esto es una asignacion en Python"
type(variable)
```

```python
# En Java, una vez asignado un tipo de dato, 
# la variable solo puede recibir valores de ese tipo

# En Python, la asignacion de valores es en tiempo de ejecucion

variable = 10
type(10)
```
