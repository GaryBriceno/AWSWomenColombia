

## Logica Booleana

El algebra booleana fue desarrollada por George Boole, que es  la base para la logica detras de las computadores y los lenguajes de programación.

Esta construida alrededor de los valores de verdad de las expresiones u objetos (Estos pueden ser **True** o **False**) y usa como base los operadores lógicos **AND**, **OR** y **NOT**.

Referencia https://realpython.com/python-or-operator/#boolean-logic

```python
# Expresion booleana: Tiene un valor de verdad, de verdadero o falso.
5 < 6
```

```python
3 == 4
```

```python
# Los operadores sirven para trabajar con dos o mas expresiones boleanas.
(5 < 6) or (3 == 4)
```

Tabla de Verdad de la expresion **OR**

| expr1       | expr2       | expr1 or expr2 |
| ----------- | ----------- | -------------- |
| True        | True        | True           |
| True        | False       | True           |
| False       | True        | True           |
| False       | False       | False          |


Tabla de Verdad de la expresion **AND**

| expr1       | expr2       | expr1 and expr2 |
| ----------- | ----------- | --------------- |
| True        | True        | True            |
| True        | False       | False           |
| False       | True        | False           |
| False       | False       | False           |


```python
(3 == 6)
```
