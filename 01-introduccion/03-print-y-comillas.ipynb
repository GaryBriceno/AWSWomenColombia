## Uso del print


* Uso del print
* Uso de las comillas simples y dobles
* Uso de las comillas triples
* Secuencias de Escape



```python
# Uso de la funcion print, con comillas dobles
print("Hello Word!")
```

```python
print('Hello Word!')
```

```python
# Uso de print con numeros
print(10)
```

```python
# Impresion de valores, separados por coma
print('Welcome', 'to', 'Python')
```

```python
# Impresion de saltos de linea
print('Welcome\nto\n\nPython!')
```


Secuencias de Escape

* \n : Inserta una nueva linea.
* \t : Inserta espacios, similar a un tab.
* \\\ : Inserta un caracter back slash en la cadena.
* \\' : Inserta una comilla simple en la cadena.
* \\" : Inserta una comilla doble en la cadena 



```python
# Impresion de comilla simple
print("O'Reilly")
```

```python
print('O\'Reilly')
```
